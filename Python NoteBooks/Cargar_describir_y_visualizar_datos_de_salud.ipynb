# Cargar describir y visualizar datos de salud


## Cargar librerías

```python
# Cargar librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Cargar funciones

* La función f_cargar_datos(n=200) crea un conjunto de 200 datos aleatorios con cinco variables numéricas y variables categóricas.
* La función f_describir_datos_numéricos(df) describe estadísticamente los datos numéricos.
* La función *f_visualizar_datos(df)* presenta visualmente la frecuencia de la condiciu´pon de saluda de los registros.

```python
# Crear funciones
def f_cargar_datos(n=200):
    """
    Genera un dataset simulado de salud
    con 5 variables numéricas y 2 categóricas.
    """

    np.random.seed(2026)

    datos = pd.DataFrame({
        "edad": np.random.randint(18, 80, n),
        "presion_sistolica": np.random.normal(120, 15, n),
        "glucosa": np.random.normal(100, 25, n),
        "colesterol": np.random.normal(200, 30, n),
        "imc": np.random.normal(27, 4, n),
        "sexo": np.random.choice(["Masculino", "Femenino"], n),
        "diagnostico": np.random.choice(
            ["Sano", "Hipertensión", "Diabetes"], n
        )
    })

    return datos


# ==========================================
# 3. Función para describir datos numéricos
# ==========================================

def f_describir_datos_numéricos(df):
    """
    Muestra estadísticas descriptivas
    de las variables numéricas.
    """

    descripcion = df.describe()
    print("\nDescripción estadística de variables numéricas:\n")
    print(descripcion)

    return descripcion


# ==========================================
# 4. Función para visualizar datos
# ==========================================

def f_visualizar_datos(df):
    """
    Genera un gráfico de frecuencia
    de la variable categórica 'diagnostico'
    """

    plt.figure()
    sns.countplot(x="diagnostico", data=df)
    plt.title("Distribución de Diagnóstico")
    plt.xlabel("Diagnóstico")
    plt.ylabel("Frecuencia")
    plt.show()
```

## Ejecutar funciones

```python
# Configuración visual
sns.set(style="whitegrid")


# Cargar datos
datos = f_cargar_datos()

# Mostrar primeras filas
print("\nPrimeras filas del dataset:\n")
print(datos.head())

# Descripción estadística
f_describir_datos_numéricos(datos)

# Visualización
f_visualizar_datos(datos)
```

# Interpretación del caso

Se construye un conjunto de datos estructurados con cinco variables numéricas y dos de tipo categórico. El contexto de los datos es del sector salud y contiene un total de 200 registros.

La imnple,mentación del caso se hace a través de las carga de librerías necesarias y la construcción de funciones personalizadas que se ejecutan al final del mismo.
