# Contexto de los datos

Este caso de estudio trata con datos que caracterizan y definen la calidad de vino blanco, son alrededor de 4898 registros y 25 variables.

Las variables independientes son:

* *fixed_acidity*: Acidez fija del vino. Representa principalmente ácidos no volátiles, como el ácido tartárico, que contribuyen al sabor ácido y estructura del vino.
* *volatile_acidity*: Acidez volátil. Corresponde principalmente al ácido acético presente en el vino; niveles altos pueden producir sabor avinagrado y disminuir la calidad.
* *citric_acid*: Ácido cítrico presente en el vino. Puede aportar frescura y mejorar el equilibrio sensorial.
* *residual_sugar*: Azúcar residual restante después de la fermentación. Influye en el dulzor percibido del vino.
* *chlorides*: Concentración de cloruros o sales. Valores altos pueden afectar negativamente el sabor del vino.
* *free_sulfur_dioxide*: Dióxido de azufre libre. Actúa como conservador y antioxidante protegiendo al vino contra bacterias y oxidación.
* *total_sulfur_dioxide*: Dióxido de azufre total presente en el vino. Incluye formas libres y combinadas del compuesto.
* *density*: Densidad del vino. Está relacionada con el contenido de azúcar y alcohol.
* *pH*: Nivel de acidez o alcalinidad del vino. Valores menores indican mayor acidez.
* *sulphates*: Concentración de sulfatos. Pueden contribuir a la estabilidad microbiológica y sensorial del vino.
* *alcohol*: Porcentaje de alcohol presente en el vino. Es uno de los factores más influyentes en percepción de calidad.
* *indice_alcohol_acidez*: Variable derivada que representa la relación entre alcohol y acidez fija. Busca modelar equilibrio estructural del vino.
* *balance_azucar_acidez*: Relación entre azúcar residual y acidez volátil. Intenta capturar balance entre dulzor y acidez.
* *indice_mineralidad*: Índice derivado a partir de minerales y sales presentes, asociado con características minerales del vino.
* *potencial_alcoholico*: Variable derivada relacionada con la capacidad potencial de generación alcohólica del vino.
* *relacion_SO2*: Relación entre dióxido de azufre libre y total. Evalúa proporción de protección activa del vino.
* *alcohol_2*: Término polinomial cuadrático del alcohol. Permite modelar relaciones no lineales del alcohol con la calidad.
* *density_2*: Término cuadrático de densidad para modelar curvaturas no lineales.
* *sugar_2*: Término cuadrático del azúcar residual. Captura posibles efectos curvos entre dulzor y calidad.
* *acidez_2*: Término cuadrático de acidez fija utilizado para modelar relaciones no lineales.
* *alcohol_sulphates*: Variable de interacción entre alcohol y sulfatos. Representa posibles efectos combinados sobre calidad.
* *density_sugar*: Variable de interacción entre densidad y azúcar residual. Modela relaciones conjuntas entre ambas variables.
* *acidez_pH*: Interacción entre acidez fija y pH. Busca representar equilibrio químico del vino.
* *alcohol_acidez_volatil*:	Interacción entre alcohol y acidez volátil. Permite modelar posibles efectos combinados sobre la percepción sensorial.

La variable dependiente es calidad de vino de manera cuantitativa:
* *quality*: Representa la calidad del vino evaluada generalmente en escala ordinal por expertos catadores.


Los datos originales son valores cuantitativos y se normalizan de manera inicial para que todos los modelos se entrenen con datos estandarizados con media igual a cero $\mu=0$ y desviación estándar igual a 1 $\sigma = 1$;  $z= \frac {x_i - \mu}{\sigma}$.
